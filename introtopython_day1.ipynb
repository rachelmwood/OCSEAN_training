# Variables and Data Types

A **variable** is a symbolic name that refers to an object. It acts as a container for storing values.

We assign values to variables by using `=`. For example:







```python
x = 5
```

The code above assigns the value `5` to the variable `x`.

To see what is stored in variable `x` we can print the variable.

```python
print(x)
```

## Rules for naming variables

* Variable names **can** include letters (both uppercase and lowercase), digits, and underscores (_)
  E.g. `my_var` or `var_2`
* They **cannot** start with a digit. E.g. `2var` (starts with a digit) or `my-var` (contains a hyphen).
* Variable names are **case-sensitive**, meaning `myVariable`, `MyVariable`, and `MYVARIABLE` would be considered different variables.




## Data Types

We will cover three different data types: numbers, strings and booleans.


### Numeric data

Numbers in Python exist as intergers (whole numbers) or floating-point numbers (numbers with decimal places)

Description                       | Type     | Example
----------------------------------|----------|----------------
integers or whole numbers         | int      | a = int(12)
floating-point numbers or decimals          | float    | b = float(7.3)


You can use the `type()` function to check the data type of a variable.



```python
a = 12
b = 7.3
print("A:", a, type(a))
print("B:", b, type(b))
```

### String variables

String variables refer to text. They are also typed using quotation marks e.g. `"string"`

Description                       | Type     | Example
----------------------------------|----------|----------------
textual data (strings)            | str      | a = "Jean Golding"


```python
name = "Python"
print(name, type(name))
```

It’s important that when writing numbers, you do not put quotation marks around them, otherwise they will be recognized by Python as strings. There is a difference between `3.14159` and `"3.14159"`.




```python
pi = 3.14159
print("pi:", pi, type(pi))
```

```python
pi = "3.14159"
print("pi:", pi, type(pi))
```

### Boolean Types

Boolean variables can only hold one of two values: true or false. We can use them to understand the relationship between two variables e.g:


```python
a = 23
b = 2
c = (a == b)
print("c:", c, type(c))
```

### Converting data types

Data types determine what operations can be performed on the data.
You can convert between different data types using built-in functions like `int()`, `float()` and `str()`.

# Operators

Operators are essential for performing various operations on variables and values. For example, multipling two numbers or comparing them to know which one is greater.

## Arithmetic Operators

Arithmetic operators can be used with numerical variables such as `int` and `float`. Some of them are:

Description                 | Operator | Example
----------------------------|----------|----------------
Addition                    | +        | a + b
Subtraction                 | -        | a - b
Multiplication              | *        | a * b
Division                    | /        | a / b
Exponentiation              | **       | a ** b

```python
a = 9
b = 2
print('Subtraction:', a - b)
print('Division:', a / b)
print('Power:', a ** b)
```

## Comparison Operators

We can compare two variables (or a variable and a value) using comparison operators. They are applicable to numerical variables, strings and booleans.

Description                 | Operator | Example
----------------------------|----------|----------------
Equal To                    | ==       | a == b
Not Equal To                | !=       | a != b
Greater Than                | >        | a > b
Less Than                   | <        | a < b
Greater Than or Equal To    | >=       | a >= b
Less Than or Equal To       | <=       | a <= b


```python
a = 5
b = 2
print("a is equal to b:", a == b)
print("a is less than b:", a > b)
print("a is not equal to b:", a != b)
print("a is greater than b:", a < b)
```

# Lists

Until now all the variables we have used have contained a single value. It’s very common to want to refer to collections of data.

A list is a data type that stores an ordered sequence of elements. It can be created using square brackets `[]`, with elements separated by commas `,`.

```python
my_list = ["cat", "dog", "horse"]

print(my_list)
```

A list can also have zero items in it:

```python
my_list = []

print(my_list)
```

It can also have a mixture of data types

```python
my_list = [5, 34.6, "Hello", -6, False]

print(my_list)
```

## Indexing

Indexing allows us to extract specific elements of the list, using square brackets `[]`.

We refer to elements in the list by the order in which they were added to it.

In Python we start counting from **zero** `0`, called zero-indexing.
For example:

```python
my_list = ["cat", "dog", "horse", "mouse"]

print(my_list[0])
```

`my_list[0]` extracts the first value in the list: `"cat"`

### Reverse Indexing

If we want to access the **last** element in the list we can use negative numbers. In the case of reverse indexing we start from `-1`.

```python
my_list = ["cat", "dog", "horse", "mouse"]

print(my_list[-1])
```

**Exercise:** Use indexing and reverse indexing to practise extracting values from the list above.

### Slicing

You can also grab subsections of a list. This is called slicing.

We use the same zero-indexing method above (`0` = the first item in the list). However Python will print list elements from the starting index up to, **but not including**, the end index. For example:

```python
my_list = ["cat", "dog", "horse", "mouse"]

print(my_list[0:3])
```

You see that it printed `['cat', 'dog', 'horse']` which is:
* Index `0` (`"cat"`)
* Index `1`(`"dog"`)
* Index `2` (`"horse"`)

Index `3` (`"mouse"`) was not printed.



### Adding elements to list

You can add items to the end of your list by using the `append` function.

The `append` function is used slightly differently to other functions that we have used so far like `print()`.

To use `append`, we give the name of our list (`my_list`), followed by a dot `.` and then the name of our function (`append`). See below:


```python
my_list = ["cat", "dog", "horse", "mouse"]

my_list.append("duck")

print(my_list)
```

We can remove items from our list by using the `remove` function in a similar way.

# Dictionaries

Sometimes you want to store even more variables but access them using more complex relationships. A dictionary lets you store variables and access them using a key.

Dictionaries are created using curly brackets `{}`.


```python
sounds = {"cat": "meow", "dog": "woof", "horse": "neigh"}
```

Here we have created a dictionary called `sounds`.

Each item in the dictionary comprises a *key* (`"cat"`) and *value* (`"meow"`), seperated by a colon `:` The *value* is the real data that we want to keep hold of and the *key* is how we can get at the data we want.

We can extract the value using the key:

```python
sounds = {"cat": "meow", "dog": "woof", "horse": "neigh"}

cat_sound = sounds["cat"]

print(cat_sound)
```

Notice how when we extract a value from the dictionary, we use square brackets `[]` similar to indexing.

## Adding new data into dictionaries

You can add new values to a dictionary using the syntax below

```python
sounds = {
    "cat": "meow",
    "dog": "woof",
    "horse": "neigh"
}

sounds["cow"] = "moo"

print(sounds)
```

Here we have added the new value `"moo"` with the key `"cow"`.
