# Control Structures
Up until now, we have only looked at code that runs one line at a time. This often leads to repeated code which isn't very efficient. To avoid this coders will use **loops** and **if statements**.

path.jpg


**If statements** allow us to treat objects differently depending on their value. For example:



```
IF  sample value > 100
        DO SOMETHING
    OTHERWISE
        DO SOMETHING ELSE
```
**Loops** let us apply the same code to multiple objects. For example:

```
LOOP FOR EACH sample IN study
    DO SOMETHING
```


## If Statements

If statements are made up of a condition for which we apply code. This condition must evaluate to `True` or `False`. If the value is `True`, the indented code will run. If the value is `False`, the value inside the code will not run.

````python
age = 20
if age >= 18:
    print("You are an adult")
````

You can also add an else clause to specify code to run if the condition is `False`:

````python
age = 10
if age >= 18:
    print("You are an adult")
else:
    print("You are not an adult")
````

For multiple conditions, you can use `elif` (else if) clauses to execute the first matching block.



````python
age = 14
if age < 13:
    print("You are a child")
elif age < 18:
    print("You are a teenager")
else:
    print("You are an adult")
````

### Ordering Options

Python will work down the list of `if`, `elif`s and else and will run the first one that matches. After the first `True` condition, it will not check any of the later options. For example:

````python
my_number = 1000

if my_number > 100:
    print(my_number, "is large")
elif my_number > 1:
    print(my_number, "is positive")
else:
    print(my_number, "negative")
````

will give a different output to this:

````python
if my_number > 1:
    print(my_number, "is positive")
elif my_number > 100:
    print(my_number, "is large")
else:
    print(my_number, "negative")
````

### Combining Conditions

Sometimes you might need to run code only if a combination of conditions are `True`. This is done with `and` or `or`. For example, if you want to check is a number is smaller than ten `(my_number < 10)` and is not equal to zero `(my_number != 0)`, you can use:

````python
if my_number < 10 and my_number != 0:
  print(my_number)
````

## For Loops
It is possible to make Python repeat certain lines of code using **loops**.

Imagine we have strings in a list that we want to print, we could do:

````python
words = ['My', 'name', 'is', 'Python']

print(words[0])
print(words[1])
print(words[2])
print(words[3])
````

We done what we wanted to do, but the repeated code is wasteful. Instead, we can use a loop to shorten this:

````python
my_words = ['My', 'name', 'is', 'Python']

for number in my_words:
  print(number)
````

**Excercise:** Change the `my_words` list to output your own name.

The beginning of the `for` loop is composed of the following parts:

<font color="red">**`for`**</font> `word` <font color="red">**`in`**</font> `my_words` <font color="red">**`:`**</font>

This breaks down to:
- The red text shows the python syntax that stays the same for any `for` loop.
- `my_words` is the object we want to loop over (e.g a list or a dictionary).
- `word` is the name we choose for the variable for each element within the loop.

The **body** of the loop what we call the code we want to run repeatedly. This is indicated through indentation.

If we want to put code after the loop, we make sure it's not indented:

````python
my_words = ["Hello", "Python"]

for word in my_words:
    print(word)

    print("...Goodbye")
````

### Looping Over Strings


There are many things we might choose to loop over, for example strings, lists, and dictionaries.

````python
phrase = "Hello Python"

for letter in phrase:
  print(letter)
````

### Looping Over Numbers
Sometimes you want to perform a loop a set number of times. For this we can use the `range` function, which gives us every number (up to but not including) the input number starting from 0.

````python
for num in range(5):
  print(num)
````


### Looping over dictionaries

Python also allows you to loop over dictionaries.

First, let's seee what happens when we loop over a dictionary:

````python
sounds = {
    "cat": "meow",
    "dog": "woof",
    "horse": "neigh"
}

for animal in sounds:
    print(animal)
````

You can see that looping over a dictionary gives us the keys.

If we want to loop over the values, we need to include the method `values`:



````python
sounds = {
    "cat": "meow",
    "dog": "woof",
    "horse": "neigh"
}

for sound in sounds.values():
    print(sound)
````

If we want both the keys and values, then we use the method `items`. The `items` method gives us two pieces of data, first the key and then the value.

We need to give **two** variable names in the `for` loop declaration. We give the keys the name `animal` and the `values` the name sound.

````python
sounds = {
    "cat": "meow",
    "dog": "woof",
    "horse": "neigh"
}

for animal, sound in sounds.items():
    print(animal, "goes", sound)
````

# Functions

Functions are another way to avoid repeating your code, as they allow you to package code into reusable components.

We've already used some functions already, such as `print()` and `append()`.

Here we're going learn how to write our own functions.

## Defining a Function
An example of code we might want to repeat a lot is convert between units.

Say we want to convert from degrees Celcius to degrees Farenheit. Without using a function we would write something like this:




````python
degree_celcius = 30

degree_farenheit = degree_celcius * 1.8 + 32
print(degree_farenheit, "in degrees Farenheit")
````

We can break this down into three parts:

- The **set-up** where we define `degree_celcius`
- The **data-processing** section where we read our inputs and create an output
- The **output** section where we print our result to the screen

We want to create code where the data processing step will work for any variable we give it.

Turning this into a function, we get:

````python
def celcius_to_farenheit(degrees):
  new_degrees = degrees * 1.8 + 32
  return new_degrees
````

Simular to for loops, this has a structure we need to follow. The first line in defining the function is

<font color="red">**`def`**</font> `celcius_to_farenheit` <font color="red">**`(`**</font> `degrees` <font color="red">**`)`**</font> <font color="red">**`:`**</font>

This is made up of:
- The red text shows the Python syntax that is the same for any function
- `celcius_to_farenheit` is the name of the function
- `degrees` is the input to the function. A function can also have 0 or more than one input.

Everything that is indented after this is called the body of the function. The final line tells us what we want to send back after the function is done.

## Calling a function

Calling a function just means we run the code inside the function. We do it like this:

````python
def celcius_to_farenheit(degrees):
  new_degrees = degrees * 1.8 + 32
  return new_degrees

degree_celcius = 30

degree_farenheit = celcius_to_farenheit(degree_celcius)

print(degree_farenheit, "in degrees Farenheit")
````

## Importing Functions from Modules

While some functions are ready for you to use straight away, some are part of **modules**, which need to be imported.

An example of a module is the `math` module. We load it like this:

````python
import math
````

We can then call the function by typing the name of the function, then a dot and the name of the function:

````python
import math
print(math.sqrt(25))
````

You can think of this as saying “from the `math` module that I've just imported, get the `sqrt` function and call it”.

Alternatively, if you want to grab a specific function out of a module so that you can call it without specifying the module name, you can do:



````python
from math import sqrt

print(sqrt(25))
````
