### Headers
One of the first things we notice is that the first few lines of `city_pop.csv` are not the actual data (what we call a header). We can skip these like this:

````python
pop = pd.read_csv(
    'city_pop.csv',
    skiprows=5,
)
````

Checking what this looks like now, we can see the non-data lines are missing, but there is still an issue with how the rows are processed:

````python
pop
````

### Seperators

By default, the function assumes the values are seperated by a comma (,). Since our data is seperated by a semicolon (;), we need to specify this when calling our function:

````python
pop = pd.read_csv(
    'city_pop.csv',
    skiprows=5,
    sep=";",  # Add this
)
````

We can look at our data and see that it is now in a `pandas DataFrame` format that we can work with:

````python
pop
````

### Identifying missing data

Reading the descriptive header of our data file we see that a value of `-1` signifies a missing reading so we should mark those too. This can be done using the `na_values` argument:

````python
pd.read_csv(
    "city_pop.csv",
    skiprows=5,
    sep=";",
    na_values="-1",  # Add this
)
````

# Querying Data
Now that we know how to load our data into python, we want to know how to extract the information we want.

## Accessing Columns
Going back to the `students` data, we might want to extract a specific column, e.g. the reading scores. As with lists and dictionaries, we access these with square brackets:

````python
students['reading score']
````

In a similar way, if we wanted to look at all scores, we can use a list of columns:

````python
students[['math score', 'reading score', 'writing score']]
````

## Accessing Rows
For this, we can use the `.loc` (short for location) attribute like this:

````python
students.loc[2]
````

We can also use `loc` to access a single column of a row, using the column name:

````python
students.loc[2, 'gender']
````

# Manipulating Data
Now that we know how to refer to individual columns, we can start asking questions about the data therein. If you've worked with columns of data in Excel for example, you've probably come across the `SUM()` and `AVERAGE()` functions to summarise data. We can do the same thing in pandas by calling the `sum()` or `mean()` methods on a column:



````python
students['math score'].sum()
````

````python
students['math score'].mean()
````

There are also many other functions you can use on columns ([you can find a list here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html)), but some include:
- `max()` - gets the maximum value of a column
- `idxmax()` - gets the location of the maximum value of a column
- `min()` and `idxmin()` - same but for finding the minimum
- `count()` - gets the number of non-NA/null observations in a column

## Creating New Columns
In some situations we instead want to manipulate a column to create a new column. For example, the scores are given out of 100, but we might want to change them to be out of 10:

````python
students['writing score']/10
````

You can do any mathematical operation that Python supports, such as `+`, `-` and `/`.

## Combining Columns

As well as operating on individual columns, you can combine together multiple columns. Any operation you do between two columns will be done row-wise, that is adding two columns will add together the two values from the first row of each, then the second row from each etc.

Say we want to find the average score for each student, we could write:

````python
(students['math score'] + students['reading score'] + students['writing score'])/3
````

So far we've just been outputing the new columns we created, but we might want to add them to our table. To add the average score to our `students` `DataFrame`, we can do this:

````python
students['average score'] = (students['math score'] + students['reading score'] + students['writing score'])/3
students
````

# Filtering Data

Sometimes we don’t want to have to deal with entire columns at once, we might only want to grab a subset of the data and look in just that part.

For example in the `students` data we might want to just look at the scores for students who completed the test preparation course.

In Pandas there are two steps to asking a question like this.

1. create a filter which describes the question you want to ask
2. apply that filter to the data to get just the bits you are interested in

For our example we create the filter like this:

````python
prep_filter = students['test preparation course'] == 'completed'
prep_filter
````

We can use this to filter the DataFrame as a whole to select only students who have `completed` the `test preparation course`:

````python
students[prep_filter]
````

Notice that it now says that the table only has 358 rows, down from 1000. However, the index has been maintained. This is because the row labels are connected to the row, they’re not just row numbers.

Instead of creating and naming a filter, you can also do this in one step:

````python
students[students['test preparation course'] == "completed"]
````

As well as filtering with the `==` operator (which only checks for exact matches), you can do other types of comparisons. Any of the standard Python comparisons will work (i.e. `==`, `!=`, `<`, `<=`, `>`, `>=`).

### Combining filters

You can apply multiple filters at once using the `&` operator.

For example, if we wanted to filter for students who had `completed` the `test preparation course` and achieved an `average score` greater than or equal to (`>=`) `50`:

````python
students[(students['test preparation course'] == "completed") & (students['average score'] >= 50)]
````

We seperate each filter using `()`

If you want to do an “or” operation, then instead of `&` you can use `|`.

## Plotting *relationships* between variables in the data set

We’ll start with looking at how we can visualise how data are related to each other. This is showing how a change in one variable affects another variable. For example it could be scatter plot of how age and weight are related or a line plot of the change in temperature over time.


seaborn offers a generic interface for doing this called `relplot()`.

You use `relplot()` by passing it the DataFrame you want to explore and the two dimensions of the table that you want to compare. This only makes sense for dimensions that are numerical.

The `data` argument should be the entire DataFrame and the `x` and `y` arguments should be the names of the columns to plot. For this example we will use `"math score"` and `"writing score"`:

````python
sns.relplot(
    data=students,
    x="math score",
    y="writing score",
)
````

By default it creates a scatter plot and uses the column names as axis labels. To set the axis labels to something more useful, you can call `set()` on the object returned by `relplot()`:

````python
sns.relplot(
    data=students,
    x="math score",
    y="writing score",
).set(
    xlabel="Math Score (%)",
    ylabel="Writing Score (%)",
)
````

seaborn provides a bunch of different dimensions onto which you can map your data:

- `x=`: numerical variable
- `y=`: numerical variable
- `hue=`: assign a different colour depending on the value.
- `size=`: vary the size of the marker depending on the value
- `style=`: use a different marker type (e.g. ▲, ◆, ●, ◼) for the variable. Only makes sense for discrete variables.
- `row=` and `col=`: for discrete variables, draw a separate subplot along a row or column for each category

Here is an example of using 4 different dimensions for different variables

````python
sns.relplot(
    data=students,
    x="math score",
    y="writing score",
    hue= "gender",
    style= "parental level of education",
).set(
    xlabel="Math Score (%)",
    ylabel="Writing Score (%)",
)
````

Plotting lots of dimensions can make a graph hard to interpret. Therefore it is recommended to keep the number of variables plotted at once small. You can also use the plotting dimensions redundantly, i.e. using both the marker style and the hue for the same data variable.

#Plotting the distribution of variables

This is usually about *how many* of each value you have and how they are spread out. For example it could be a histogram of the spread of marks for a class of students or the ages of people in a country. As with the x and y axes for relplot, this only makes sense for variables that are numerical.

seaborn provides a function for investigating the distribution of a variable called `displot`, which plots a histogram.

For example to plot the distribution of `average score`:

````python
sns.displot(
    data=students,
    x="average score",
)
````

We can also display different dimensions of our data.

Here we can use `col=` to plot two seperate histograms for each category in the `gender` column:

````python
sns.displot(
    data=students,
    x="average score",
    col = "gender",
)
````

Plotting things side by side can make them hard to compare. We can overlay them using `hue=`

````python
sns.displot(
    data=students,
    x="average score",
    hue = "gender",
)
````

And we can also change from a bars to a line:

````python
sns.displot(
    data=students,
    x="average score",
    hue = "gender",
    kind="kde",
)
````

## Seeing how the data varies by category

When you have categorical variables in your data, you usually want to compare between the categories.

We can use the function `catplot` for this.

We will plot a boxplot to see how `average score` varies by `parental level of education`. We use the argument `kind=` to specify a boxplot:

````python
sns.catplot(
    data=students,
    x="gender",
    y="average score",
    kind = "box"
)
````

We can also reorder the groups on the x axis. We do this using the `order` parameter:

````python
sns.catplot(
    data=students,
    x="gender",
    y="average score",
    kind = "box",
    order=["male", "female"],
)
````
